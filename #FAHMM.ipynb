```python
def theorymselow(P,sta,id,sp,ap,sigma):
    #sta=np.linalg.eig(P.T)[1]
    #sta=sta[:,0]/np.sum(sta[:,0])
    coef1=sta[0]*P[0,0]*sigma+sta[1]*P[1,1]*sigma
    coef2=sta[0]*P[0,1]+sta[1]*P[1,0]
    coef2*=id**2
    A=pow(0.5,ap-sp-2)*(1-pow(0.5,sp))**2
    B=5*(3*sp+4-pow(0.5,sp-2)*(2*sp+1))
    print(coef2)
    return np.sqrt(coef1+coef2*(A+B))
```

```python
def theorymseup(P,sta,id,sp,ap,sigma):
    #sta=np.linalg.eig(P.T)[1]
    #sta=sta[:,0]/np.sum(sta[:,0])
    coef1=sigma
    coef2=sta[0]*sta[1]**2+sta[1]*sta[0]**2
    coef2*=id**2
    A=pow(0.5,ap-sp-2)*(1-pow(0.5,sp))**2
    B=5*(3*sp+4-pow(0.5,sp-2)*(2*sp+1))
    print(coef2)
    return np.sqrt(coef1+coef2*(A+B))
```
